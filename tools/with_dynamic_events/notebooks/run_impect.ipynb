```python
import json
import sys

import pandas as pd
from pathlib import Path
from io import BytesIO

# Path setup to import from the project root
project_root = Path().resolve().parent.parent.parent
sys.path.insert(0, str(project_root))

from event_synchronization.with_dynamic_events.impect import ImpectSyncDynamicEventsManager
from event_synchronization.events_utils.impect import ImpectEvents
from skillcorner.client import SkillcornerClient

skc_client = SkillcornerClient(username='USERNAME', password='PASSWORD')
```

# Load data

```python
data_dir = 'path/to/your/skc_dynamic_event/and/wyscout/match/folder/'
```

```python
# load match data with SKC match_id
MATCH_ID = 0
match_data = skc_client.get_match(MATCH_ID)

# load Skillcorner dynamic events
api_skc_events = skc_client.get_dynamic_events(
                match_id=MATCH_ID, params={'file_format': 'csv', 'ignore_dynamic_events_check': False}
            )
skc_events = pd.read_csv(BytesIO(api_skc_events))

# load impect match_data
impect_match_data_path = data_dir + 'impect_match_data.json'
with open(impect_match_data_path) as f:
    impect_match_data = json.load(f)

# load impect events
raw_events_path = data_dir + 'raw_impect_events.json'
with open(raw_events_path) as f:
    raw_impect_events = json.load(f)
```

# Run DynamicEvent Synchronization Process

```python
impect_events_manager = ImpectEvents(raw_impect_events, impect_match_data, match_data)
impect_manager = ImpectSyncDynamicEventsManager(skc_events, raw_impect_events, impect_events_manager)

# apply the synchronization process
skc_events_mapping, impect_events_mapping = impect_manager.run()
```
